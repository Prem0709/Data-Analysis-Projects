### The **LinkedInJobs_MLDataset** contains job listings data sourced from LinkedIn, providing insights into job postings, salaries, company details, and application engagement. Below is a detailed breakdown of the dataset's key columns:

### **Dataset Overview:**
- **Total Rows:** 33,246 job postings  
- **Total Columns:** 24  

### **Column Descriptions:**
1. **Co_Nm** (Company Name) â€“ Name of the company posting the job.  
2. **Co_Pg_Lstd** (Company Page Listed) â€“ Indicates if the company has a LinkedIn page (`True/False`).  
3. **Emp_Cnt** (Employee Count) â€“ Approximate number of employees in the company.  
4. **Flw_Cnt** (Follower Count) â€“ Number of LinkedIn followers the company has.  
5. **Job_Ttl** (Job Title) â€“ Title of the job listing.  
6. **Job_Desc** (Job Description) â€“ Full job description text.  
7. **Is_Supvsr** (Is Supervisor) â€“ Indicates if the job is a supervisory role (`True/False`).  
8. **max_sal** (Maximum Salary) â€“ Highest estimated salary for the position.  
9. **med_sal** (Median Salary) â€“ Median estimated salary for the position.  
10. **min_sal** (Minimum Salary) â€“ Lowest estimated salary for the position.  
11. **py_prd** (Pay Period) â€“ The salary payment frequency (e.g., yearly, monthly).  
12. **py_lstd** (Pay Listed) â€“ Indicates whether salary information is listed (`True/False`).  
13. **wrk_typ** (Work Type) â€“ Work setting (e.g., Full-time, Part-time, Contract).  
14. **loc** (Location) â€“ City and state of the job posting.  
15. **st_code** (State Code) â€“ Abbreviation of the jobâ€™s state location.  
16. **is_remote** (Is Remote) â€“ Indicates if the job is remote (`0=Onsite, 1=Remote`).  
17. **views** â€“ Number of views the job posting received.  
18. **app_typ** (Application Type) â€“ Indicates whether the application is done onsite or offsite.  
19. **app_is_off** (Application Is Off) â€“ Indicates if applications are closed (`True/False`).  
20. **xp_lvl** (Experience Level) â€“ Required experience level (Entry, Mid-Senior, Director, etc.).  
21. **domain** â€“ The domain or website where the job was posted.  
22. **has_post_domain** â€“ Indicates if a domain was listed (`True/False`).  
23. **is_sponsored** â€“ Indicates if the job post is sponsored (`0=Not Sponsored, 1=Sponsored`).  
24. **base_comp** â€“ Indicator if base compensation details are provided (`0/1`).  

### **Potential Uses:**
- **Job Market Trends** â€“ Identify top hiring companies and in-demand job roles.  
- **Salary Analysis** â€“ Compare salary ranges across job titles, experience levels, and locations.  
- **Remote Work Trends** â€“ Analyze the prevalence and salaries of remote jobs.  
- **Engagement Insights** â€“ Understand which job postings attract more views and applications.

- The **executive level** has the largest spread, with some extreme outliers in the higher salary range.

````python
from scipy.stats import skew

# Calculate skewness for the salary column
skewness_value = skew(df['med_sal'], nan_policy='omit')

print(f"Skewness of salary distribution: {skewness_value}")
````

To numerically detect skewness in your salary distribution, we can compute the **skewness statistic** using **SciPy**. Skewness measures the asymmetry of the data:

- **Positive skew (> 0):** Right-skewed (long tail on the right, higher salaries pulling the mean up).
- **Negative skew (< 0):** Left-skewed (long tail on the left, uncommon in salary data).
- **Zero skew (â‰ˆ 0):** Symmetric distribution.

### **Interpreting Results:**
- If `skewness_value` is **significantly positive**, salaries are **right-skewed**, which aligns with your boxplot.
- If close to **zero**, salaries are **normally distributed**.
- If negative, salaries are **left-skewed**, which is rare in salary distributions.

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Plot original salary distribution
plt.figure(figsize=(12,5))
sns.histplot(df['med_sal'], bins=50, kde=True)
plt.title("Original Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()

# Apply log transformation
df['log_med_sal'] = np.log1p(df['med_sal'])  # log1p to handle zero values

# Plot transformed salary distribution
plt.figure(figsize=(12,5))
sns.histplot(df['log_med_sal'], bins=50, kde=True, color='orange')
plt.title("Log-Transformed Salary Distribution")
plt.xlabel("Log Salary")
plt.ylabel("Frequency")
plt.show()
````
